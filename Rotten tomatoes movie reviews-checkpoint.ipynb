```python
from bs4 import BeautifulSoup  
import urllib3

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning) #suppress warnings
http = urllib3.PoolManager()

url = 'https://editorial.rottentomatoes.com/guide/best-movies-of-2018/'
response = http.request('GET', url) #GETS the HTML contents from a webpage

#soup = BeautifulSoup(response.data)
 
soup = BeautifulSoup(response.data,'html.parser' )
print(soup.prettify())

```

```python
 
[type(item) for item in list(soup.children)] #navigable string: bit of text within a tag
```

```python
list(soup.children) #shows children tags
```

```python
a =soup.find(id='row-index-197')
a.find(class_ ='countdown-index-resposive').get_text()
x = a.find('h2')
x.find('a').get_text() #Movie title
a.find(class_ = 'subtle start-year').get_text() #Movie Year
a.find(class_ = 'col-sm-20 col-full-xs').find('col-sm-20 col-full-xs')



soup.find(class_ = 'info countdown-adjusted-score').get_text() ##The Adjusted Score comes from a weighted formula (Bayesian) that we use that accounts for variation in the number of reviews per movie."
soup.find(class_ = 'info critics-consensus').get_text() #critics consensus

#cast
soup.find(class_ = 'info cast').get_text()

#director
soup.find(class_ = 'info director').get_text()

#movie synopsis
z =soup.find(class_ = 'info synopsis').find('a').get('href') #getting link for main synopsis on another page
z

re = http.request('GET', z)
so = BeautifulSoup(re.data ,'html.parser' )
so.find(class_ = 'movie_synopsis clamp clamp-6').get_text() #final synopsis

so.find(id = 'reviews').get_text() #final synopsis
so.find(class_ = 'review_quote').get_text() #final synopsis



 
```

```python
#rank of movie
rank = [i.get_text() for i in soup.find_all(class_ ='countdown-index-resposive') ]
rank

#title of movie
title = [i.get_text() for i in soup.find_all(class_ ='article_movie_title') ]
title

#adjusted score of movie
adj_score = [i.get_text() for i in soup.find_all(class_ ='info countdown-adjusted-score') ]
adj_score

#critics_consensus of movie
critics_consensus = [i.get_text() for i in soup.find_all(class_ ='info critics-consensus') ]
critics_consensus

#info synopsis
info_1 = [i.find('a').get('href') for i in soup.find_all(class_ ='info synopsis') ]
info_2 = [http.request('GET', z) for z in info_1 ]
info_3 = [BeautifulSoup(re.data ,'html.parser' ) for re in info_2]
info_synopsis = [x.find(class_ = 'movie_synopsis clamp clamp-6').get_text() for x in info_3]

info_synopsis
```

```python
#cast
cast =  [i.get_text() for i in soup.find_all(class_ ='info cast') ]
cast

#director
director =  [i.get_text() for i in soup.find_all(class_ ='info director') ] 
director
```

```python
import pandas as pd
df= pd.DataFrame({"Rank":rank,"Title" :title,"Adjusted_score":adj_score,"Critics_consensus":critics_consensus,"Info_synopsis":info_synopsis,"Cast" :cast,"Director" : director})
df
```
